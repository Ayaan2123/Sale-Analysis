```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("Diwali_Sales.csv",encoding="unicode_escape")
df.head()
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.describe
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

```python
ax = sns.countplot(x = 'Gender',data = df, hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, hue='Gender')
```

```python
ax = sns.countplot(data = df, x = 'Age Group')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders', palette='cool')
```

```python

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount',palette="Greens")
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status',hue='Gender')

sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation',palette='cool',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(25,5)})
sns.countplot(data = df, x = 'Product_Category', hue='Gender')
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
